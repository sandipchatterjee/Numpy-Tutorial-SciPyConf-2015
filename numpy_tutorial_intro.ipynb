## Array Operations

### default behavior is to perform operations elementwise

```python
# Simple array math

a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
print(a+b)
print(a*b)
print(a**b)
```

```python
# Broadcasting

# for example, add a scalar to an array, and the scalar will be "broadcast" across a 1D array

a + 1
```

```python
a = np.arange(21)
a
```

```python
# multiply by a scalar, then divide by a scalar

a = np.arange(21) * 2 * np.pi/21
a
```

```python
b = np.sin(a)
plt.plot(a,b)
plt.show()
```

## matplotlib interlude

```python
# Plot array values vs. array indices

x = np.linspace(0, 2*np.pi, 50)

print(x)      # array
print(len(x)) # number of array indices

plt.plot(np.sin(x))

plt.show()
```

```python
# Multiple arrays drawn on same plot

plt.plot(x, np.sin(x),    # x1, y1
         x, np.sin(2*x))  # x2, y2
        # ...
        # can have additional x(n), y(n) drawn on the same plot
        
plt.show()
```

```python
# Line formatting

plt.plot(x, np.sin(x), 'r-^') # specify color and shape of markers (triangles in this case)
plt.show()
```

```python
# Multiple plot groups (similar to before)

plt.plot(x, np.sin(x), 'b-o',
         x, np.sin(2*x), 'r-^')
plt.show()
```

```python
# Scatter plots

# Simple scatter plot

x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
plt.scatter(x, y)
plt.show()
```

```python
# Same result as above, but using the matplotlib 'plot' function

plt.plot(x, y, 'bo')
plt.show()
```

```python
# Colormapped scatter

x = np.random.rand(200)
y = np.random.rand(200)

size = np.random.rand(200) * 30
color = np.random.rand(200)

plt.scatter(x, y, size, color)
plt.colorbar()

plt.show()
```

```python
# subplots

x = np.array([1,2,3,2,1])
y = np.array([1,3,2,3,1])

plt.subplot(2,1,1) # arguments are rows, columns, active plot
plt.plot(x)

# activate subplot #2 (see last argument)
plt.subplot(2,1,2)
plt.plot(y)

plt.show()
```

```python
# Adding lines to a plot

# By default, previous lines are "held" on a plot

plt.plot(np.sin(x))
plt.plot(np.cos(x))
plt.show()
```

```python
# set hold(False) to erase old lines

plt.plot(np.sin(x))
plt.hold(False) # only needs to be set once per plot -- can be reversed with hold(True)
plt.plot(np.cos(x))
plt.show()
```

```python
plt.hold(True) # undo previous hold(False)

# Legends

plt.plot(np.sin(x), label='sin')
plt.plot(np.cos(x), label='cos')
plt.legend()
plt.show()
```

```python
# Legends -- define legend later

plt.plot(np.sin(x))
plt.plot(np.cos(x))
plt.legend(['sin', 'cos'])
plt.show()
```

```python
# Titles and Axis Labels

plt.plot(x, np.sin(x))
plt.xlabel('radians')
plt.ylabel('amplitude', fontsize='large')
plt.title('Sin(x)')
plt.show()
```

```python
# Display gridlines in a plot

plt.plot(x, np.sin(x))
plt.xlabel('radians')
plt.ylabel('amplitude', fontsize='large')
plt.title('Sin(x)')
plt.grid() # gridlines ON
plt.show()
```

```python
# clearing and closing plots

plt.plot(x, np.sin(x))
plt.clf() # clear plot
plt.show()

# close active plot
# plt.close()
```

### image display and colormaps

```python
from scipy.misc import lena
img = lena()
plt.imshow(img, 
           extent=[-25,25,-25,25],
            )
```

```python
# using colormaps and applying to images

from matplotlib.pyplot import cm

img = lena()

plt.imshow(img, 
           extent=[-25,25,-25,25],
           cmap = cm.bone
          )
```

```python
# add colorbar

from matplotlib.pyplot import cm

img = lena()

plt.imshow(img, 
           extent=[-25,25,-25,25],
           cmap = cm.bone
          )

plt.colorbar() 
```

```python
# Histogram with n bins

n = 30

plt.hist(np.random.randn(1000), 30)
plt.show()
```

```python
# Plot with error bars

# Assume points are known with errors in both axes

x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)
yerr = np.random.rand(10)
xerr = np.random.rand(10)/3
plt.errorbar(x, y, yerr, xerr)

plt.show()
```

```python
# Plots with error bars II: Boxplot

# Assume 5 experiments have measured some quantities

# Data creation
from numpy.random import normal
ex = [normal(i, 1+i//2, size=(500,))
      for i in np.arange(4.)]

# Plot creation
positions = np.arange(len(ex)) + 1
plt.boxplot(ex, sym='k+', notch=True, positions=positions)

# Interpretation
# Red line = median
# Notch = median @ 95% CL
# Box = [25%-75%]
# Whiskers = 1.5 * box_extent

plt.show()
```
