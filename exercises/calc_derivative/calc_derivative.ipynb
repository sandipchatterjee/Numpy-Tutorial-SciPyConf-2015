## sin(x) example from calc_derivative.py file

```python
x = np.linspace(0,2*np.pi,101)
y = np.sin(x)

x, y
```

```python
def find_derivative(x, y):
    
    ''' returns the derivative (array) of x and y,
        where x and y are both numpy arrays
    '''
    
    dy = y[1:] - y[:-1]
    dx = x[1:] - x[:-1]
    
    return dy/dx
```

```python
find_derivative(x, y)
```

```python
def find_midpoint_values(x):
    
    ''' returns the midpoint values for the array x
        (shifting the array up to the right)
    '''
    
    return (x[1:]+x[:-1])/2
```

```python
find_midpoint_values(x)
```

```python
# plot

plt.figure(figsize=(8,10))

plt.subplot(2,1,1)
plt.plot(x, y, label='sin(x)')
# plt.plot(find_midpoint_values(x), find_derivative(x,y), label='derivative', marker='o')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('sin(x) and cos(x)')
plt.legend()

plt.subplot(2,1,2)
plt.plot(x, y, label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.plot(find_midpoint_values(x), find_derivative(x,y), label='derivative', marker='o')
plt.title('Derivative of sin(x)')
plt.legend()
plt.show()
```

## integration

```python

# Trapezoidal rule integration.
avg_height = (y[1:] + y[:-1])/2.0
dx = x[1:] - x[:-1]
int_sin = np.cumsum(dx * avg_height)

# Plot our integration against -cos(x) - -cos(0)
closed_form = -np.cos(x) + np.cos(0)
plt.subplot(1,2,2)
plt.plot(x[1:], int_sin,'rx', x, closed_form,'b-')
plt.legend(('numerical', 'actual'))
plt.title(r"$\int \, \sin(x) \, dx$")
plt.show()
```
