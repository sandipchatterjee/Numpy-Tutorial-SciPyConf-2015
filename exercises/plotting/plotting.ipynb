```python
%matplotlib inline
from numpy import linspace, pi, sin, cos
from matplotlib.pyplot import (plot,
                               subplot,
                               cm,
                               imread,
                               imshow,
                               xlabel,
                               ylabel,
                               title,
                               grid,
                               axis,
                               show,
                               savefig,
                               gcf,
                               figure,
                               close,
                               tight_layout
                              )

x = linspace(0, 2*pi, 101)
s = sin(x)
c = cos(x)

img = imread('dc_metro.JPG')
```

```python
subplot(2,2,1)
plot(x, c, 'r+')
plot(x, s, 'b-')
axis([0,6,-1.0,1.0])

subplot(2,2,2)
plot(x, s)
axis([0,6,-1.0,1.0])
xlabel('radians')
ylabel('amplitude')
title('sin(x)')
grid()

subplot(2,2,3)
imshow(img,
       cmap=cm.winter,
       extent=[-10,10,-10,10])

tight_layout()

show()
```

## "bonus" (#4)

```python
ax1 = subplot(2,2,1)
plot(x, c, 'r+')
plot(x, s, 'b-')
axis([0,6,-1.0,1.0])


subplot(2,2,2, sharex=ax1, sharey=ax1)
plot(x, s)
axis([0,6,-1.0,1.0])
xlabel('radians')
ylabel('amplitude')
title('sin(x)')
grid()

subplot(2,2,3)
imshow(img,
       cmap=cm.winter,
       extent=[-10,10,-10,10])

tight_layout()

show()
```
